### Exercise 1

Write a function `columns(n)` that returns an nxn numpy array with interchanging columns of 1s and 0s:

````python
def columns(n):

    arr = np.zeros((n, n), dtype=int)
    arr[:, ::2] = 1
    return arr
````

````python
print(columns(9))
````

````python
plt.figure(figsize=(2, 2))
plt.imshow(columns(9), cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
````

### Exercise 2


Write a function `lattice(n)` that returns an nxn numpy array in which all rows and columns with odd indexes consist of 1s, and all 
other entries are 0s:

````python
def lattice(n):

    arr = np.zeros((n, n), dtype=int)
    arr[:, 1::2] = 1
    arr[1::2, :] = 1
    return arr
````

````python
print(lattice(9))
````

````python
plt.figure(figsize=(2, 2))
plt.imshow(lattice(9), cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
````

### Exercise 3

Write a function `checkers(n)` that returns an nxn numpy array with checkerboard pattern of 0s and 1s.


````python
import matplotlib.pyplot as plt
import numpy as np

def checkers(n):

    arr = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(n):
            if (i + j) % 2 == 0:
                arr[i, j] = 1
            else:
                arr[i, j] = 0
    return arr
````

````python
print(checkers(8))
````

````python
plt.figure(figsize=(2, 2))
plt.imshow(checkers(8), cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
````

### Exercise 5

Write a function `bands(n)` that returns an nxn numpy with alternating bands of 0s and 1s as in the following example:

````python
import matplotlib.pyplot as plt
import numpy as np

def bands(n):
    
    arr = np.zeros((n, n), dtype=int)
    for i in range(1,n//2 +1):
        arr[i:-i, i:-i] = i % 2
    return arr
````

````python
print(bands(9))
````

````python
plt.figure(figsize=(2, 2))
plt.imshow(bands(9), cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
````
