Operations on numpy arrays are usually much faster than on lists:

```python
from time import time
```

```python
time()
```

```python
time()
```

```python
st = time()
x = 17**1000
et = time()
print(et - st)
```

```python
from math import sin

st = time()
sins = []
for i in range(10**7):
    sins.append(sin(i))

et = time()
print(et - st)
```

```python
st = time()
sins = np.sin(np.arange(10**7))
et = time()
print(et - st)
```

## Numpy and matplotlib

```python
import matplotlib.pyplot as plt
```

```python
xlist = [0.01*i for i in range(1000)]
ylist = [ x**2 for x in xlist]
plt.plot(xlist, ylist)
plt.show()
```

```python
x = np.linspace(0, 10, 1000)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 1000)
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'b-')
plt.show()
```

## Matplotlib colormaps

```python
plt.plot([0, 0, 0], 'o', ms=50, color="olive")
plt.show()
```

```python
import matplotlib.cm as cm
```

```python
plt.plot([0, 0, 0], 'o', ms=50, color=cm.jet(0.7))
plt.show()
```

```python
x = np.linspace(0, 1, 40)
plt.figure(figsize=(14, 1))
for i in x:
    plt.plot(i, 0, 's', ms=15, color=cm.jet(i))
plt.show()
```

```python
x = np.linspace(0, 1, 40)
plt.figure(figsize=(14, 1))
for i in x:
    plt.plot(i, 0, 's', ms=15, color=cm.inferno(i))
plt.show()
```

```python
x = np.linspace(0, 1, 40)
plt.figure(figsize=(14, 1))
for i in x:
    plt.plot(i, 0, 's', ms=15, color=cm.viridis(i))
plt.show()
```

```python
x = np.linspace(0, 1, 40)
plt.figure(figsize=(14, 1))
for i in x:
    plt.plot(i, 0, 's', ms=15, color=cm.viridis_r(i))
plt.show()
```
