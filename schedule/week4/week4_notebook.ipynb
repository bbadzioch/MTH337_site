```python
from math import sin

st = time()
sins = []
for i in range(10**7):
    sins.append(sin(i))

et = time()
print(et - st)
```

```python
def columns(n):
    a = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(n):
            if j % 2 == 0:
                a[i, j] = 1
    return a
```

```python
plt.imshow(columns(10), cmap="gray")
```

```python
def columns2(n):
    a = np.zeros((n, n), dtype=int)
    for j in range(n):
        if j % 2 == 0:
            a[:, j] = 1
    return a
```

```python
columns2(10)
```

```python
def columns3(n):
    a = np.zeros((n, n), dtype=int)
    a[:, ::2] = 1
    return a
```

```python
mylist = [1, 2, 3, 4, 5, 6, 7]
```

```python
mylist[1::2]
```

```python
columns3(10)
```
