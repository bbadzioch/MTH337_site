## Random graphs

````python
def random_graph(n, d):
    """
    Returns the adjacency matrix of a random graph 
    with n vertices and with expected vertex degree d.
    """
    p = d/(n-1)
    rng = np.random.default_rng()
    a = np.zeros((n, n), dtype=int)
    selection = (rng.random((n, n)) < p)
    a[selection] = 1
    b = np.triu(a, k=1)
    adj = b + b.T
    return adj
````

### Exercise 1

The goal of this exercise is to find solutions to [New York Times Spelling Bee puzzles](https://www.nytimes.com/puzzles/spelling-bee). 

Here is a url to a [text file](https://inventwithpython.com/dictionary.txt) with a list of about 45,000 English words: 

```
https://inventwithpython.com/dictionary.txt
```


Write a function `spelling_bee(letters)` that takes as its argument a string consisting of 7 different letters, and returns a list of all words in the dictionary that satisfy the following conditions: 

1) A word is at least 4 letters long.
2) A word consists only of letters listed in `letters`
3) A words contains the first letter is `letters`.

**Note.** UB students can get [free subscription to New York Times](https://research.lib.buffalo.edu/nytimes),  which includes access to Spelling Bee puzzles. 

````python
# remove-cell
import requests

def spelling_bee(letters):
    dictionary = requests.get("https://inventwithpython.com/dictionary.txt").text.split()
    matches = []
    for w in dictionary: 
        if len(w) < 4 or letters[0] not in w:
            continue
        match = True
        for c in w: 
            if c not in letters:
                match = False
                break
        if match: 
            matches.append(w)
    return matches
````

### Exercise 2

Write a function `clean_text(txt)` that takes as its argument a string `txt` are returns a string `ctxt` such that:

- `ctxt` converts all upper case letters of `txt` to lower case.
- all non-letter characters in `txt` are replaced by the space character `" "` in `ctxt`.

````python
# remove-cell

import string 

def clean_text(txt):
    
    letters = []
    for c in txt.lower():
        if c in string.ascii_lowercase:
            letters.append(c)
        else:
            letters.append(" ")

    ctxt = "".join(letters)
    return ctxt
````

````python
s = "This is a test. 1234! Does it work?"
print(clean_text(s))
````
