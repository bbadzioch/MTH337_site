## Text cleaning

````python
import string

def clean_text(txt):
    
    letters = []
    for c in txt.lower():
        if c in string.ascii_lowercase:
            letters.append(c)
        else:
            letters.append(" ")

    ctxt = "".join(letters)
    return ctxt
````
