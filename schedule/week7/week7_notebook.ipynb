```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def random_graph(n, p):
    rng = np.random.default_rng()
    a = np.zeros((n, n), dtype=int)
    selection = (rng.random((n, n)) < p)
    a[selection] = 1
    b = np.triu(a, k=1)
    adj = b + b.T
    return adj
```

```python
adj = random_graph(20, p=0.3)
G = nx.from_numpy_array(adj)
```

```python
plt.figure(figsize=(3, 3))

pos = nx.spring_layout(G, seed=10)
#pos = nx.circular_layout(G)
colors = ["gold"]*20
colors[:10] = ["red"]*10
print(colors)
nx.draw(G, 
        pos=pos,
        node_color= colors,
        edgecolors="black",
        node_size=300,
        with_labels=True,
        font_size=8
       )
plt.title("A sample graph")
```

```python
def random_graph(n, d):
    p = d/(n-1)
    rng = np.random.default_rng()
    a = np.zeros((n, n), dtype=int)
    selection = (rng.random((n, n)) < p)
    a[selection] = 1
    b = np.triu(a, k=1)
    adj = b + b.T
    return adj
```
