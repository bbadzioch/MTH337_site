### Exercise 2

Write a function `salt_noise(img, p)` that takes as the input a 2-dimensional numpy array `img` (with colors represented by numbers between 0 and 1), and a number `p` between 0 and 1, and produces a new numpy array that adds salt and noise to `img`. Each pixel should be replaced by a noisy pixel with probability `p`.  

### Exercise 3

Write a function `sp_noise(img, p)` that takes as the input a 2-dimensional numpy array `img` (with colors represented by numbers between 0 and 1), and a number `p` between 0 and 1, and produces a new numpy array that adds salt and pepper noise to `img`. Each pixel should be replaced by a noisy pixel with probability `p`, with approximately half noisy pixels being salt noise and the other half pepper noise.  

### Exercise 4

Write a function `rvi_noise(img, p)` that takes as the input a 2-dimensional numpy array `img` (with colors represented by numbers between 0 and 1), and a number `p` between 0 and 1, and produces a new numpy array that adds random-valued impulse noise to `img`. Each pixel should be replaced by a noisy pixel with probability `p`.
