```python
def salt_noise(img, p):
    noisy_img = img.copy()
    rng = np.random.default_rng()
    selection = rng.random(img.shape)
    sel = selection < p
    noisy_img[sel] = 1
    return noisy_img
    
    
```

```python
plt.imshow(circ, cmap="gray")
plt.show()
```

```python
noisy = salt_noise(circ, p=0.1)
plt.imshow(noisy, cmap="gray")
plt.show()
```
