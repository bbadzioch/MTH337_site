## Bar plots with matplotlib

```python
exam1 = {"A": 30, "B": 60, "C": 45, "D": 10, "F": 5}
```

```python
grades = list(exam1)
grades
```

```python
nums_ex1 = list(exam1.values())
nums_ex1
```

```python
plt.bar(
    x = range(len(grades)),
    height=nums_ex1,
    color="red",
    align="edge",
    width=0.95
    
)
plt.xticks(range(len(grades)), grades)
plt.show()
```

```python
exam2 = {"A": 10, "B": 70, "C": 65, "D": 5, "F": 10}
```

```python
nums_ex2 = list(exam2.values())
nums_ex2
```

```python
plt.bar(
    x = np.arange(len(grades)) - 0.2,
    height=nums_ex1,
    color="red",
    width=0.4,
    label="exam 1"
)

plt.bar(
    x = np.arange(len(grades)) + 0.2,
    height=nums_ex2,
    color="blue",
    width=0.4,
    label="exam 2"
)
plt.legend()
plt.xticks(range(len(grades)), grades)
plt.show()
```

```python
import requests

url = "https://www.gutenberg.org/cache/epub/5658/pg5658.txt" 
txt = requests.get(url).text
print(txt[:500])
```

```python
import re
```

```python
re.findall(r"Lord", txt)
```

- `\b` = word boundary 
- `\w` = any word character (letter, digit, underscore)

Find all words consisting of 5 letters:

```python
re.findall(r"\b\w\w\w\w\w\b", txt)[:10]
```

* `{x}` = match x times
* `{x,y}` = match between x and y times
* `{x,}` = match x times or more
* `{,y}` = match y times or less 

Find all words consisting of 14 or 15 letters:

```python
re.findall(r"\b\w{14,15}\b", txt)[:10]
```

Find all words that start with "a" and end with "t":

`*` = match 0 or more times, as many times as possible.

```python
re.findall(r"\ba\w*t\b", txt)[:10]
```

`+` = match 1 or more times, as many times as possible

```python
re.findall(r"\ba\w+t\b", txt)[:10]
```

Square brackets indicate specific characters to matched:

Find all words that start with either A, B, C, and end with t.

```python
re.findall(r"\b[ABC][a-zA-Z]*t\b", txt)[:10]
```

```python
words = re.findall(r"\b[a-z]+\b", txt.lower())
words[:10]
```

Find all words consisting only of letters b, c, d, e, f, g and that are 5 letter long:

```python
re.findall(r"\b[b-g]{5}\b", txt)
```

```python
re.findall("[0-9]+", txt)[:10]
```

`\d` = matches any digit

```python
re.findall(r"\d+", txt)[:10]
```

`.` = matches any character except for the newline character.

Match all strings that start with the dollar sign $, and end with a digit. 

```python
re.findall(r"\$.*", txt)
```

- `^` matches the beginning of a string
- `$` matches the end of a string

```python
s = "a b c"
re.findall(r"\w$", s)
```

```python
re.findall(r"^\w", s)
```

```python
re.findall(r".*\d+\.\d+.*", txt)
```

`?` = the preceding character is optional

Find all words that start with "a", followed possibly by "l" and the 3 more letters:

```python
re.findall(r"\bal?\w{3}\b", txt)[:10]
```

```python
s = "$9 77 $55 $100 101"
re.findall(r"\$(\d+)", s)
```

```python
re.findall(r"\$(?:\d+)", s)
```

```python
re.findall(r"\ba(?:ll)?\w{4}\b", txt)[90:100]
```
