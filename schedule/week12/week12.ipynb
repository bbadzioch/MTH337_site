### Exercise 1. 

 Write a function `fd(num)` that for a  (positive or negative) number `num` returns the integer which is the first non-zero digit 
of that number. If there are no non-zero digits the function should return 0.

````python
def fd(num):
    num = str(num)
    num = num.replace(".", "")
    num = num.strip("+-0")
    if num.isdigit() and num:
        return int(num[0])
    else: 
        return 0
````

````python
fd(2345)
````

````python
fd(-923.45)
````

````python
fd(0.00567)
````

````python
fd(0.0)
````

### Exercise 2. 

 Write a function `ld(num)` that for a (positive or negative) number `num` returns the integer which is the last non-zero digit 
 of that number. If there are no non-zero digits the function should return 0.

````python
def ld(num):
    num = str(num)
    num = num.replace("-", "")
    num = num.replace(".", "")
    num = num.strip("0")
    if num.isdigit() and num:
        return int(num[-1])
    else:
        return 0
````

````python
ld(-123.450)
````

````python
ld(42000.00)
````

````python
ld(0.0)
````

### Exercise 3

In the Wikipedia file, each article begins with a line that starts with `**TITLE:**` followed by the title of the article. Create a list consisting of all article titles.

````python
import re
import requests

url = "https://cdn.jsdelivr.net/gh/bbadzioch/mth337_site@main/schedule/week12/wikipedia_sample.txt"

wiki = requests.get(url).text
titles = re.findall(r'\*\*TITLE:\*\*.*', wiki)
len(titles)
````

### Exercise 4

Create a list, such that each entry of the list is a single line of the Wikipedia file that contains the string "2026"

````python
import re
import requests

url = "https://cdn.jsdelivr.net/gh/bbadzioch/mth337_site@main/schedule/week12/wikipedia_sample.txt"

wiki = requests.get(url).text
lines = re.findall(r'.*2026.*', wiki)
lines
````

### Exercise 5

Create a list, such that each entry of the list is a url of a pdf file. Such a url starts with either `http://` or `https://` followed by some characters, and ends with `.pdf`

````python
import re
import requests

url = "https://cdn.jsdelivr.net/gh/bbadzioch/mth337_site@main/schedule/week12/wikipedia_sample.txt"

wiki = requests.get(url).text
pdfs = re.findall(r'https?://.+\.pdf', wiki)
pdfs
````

### Exercise 6

Create a list, such that each entry of the list is a date written in the month/day/year format. Month and day can consist of either one or two digits, year consists of between 2 and 4 digits.

````python
import re
import requests

url = "https://cdn.jsdelivr.net/gh/bbadzioch/mth337_site@main/schedule/week12/wikipedia_sample.txt"

wiki = requests.get(url).text
dates = re.findall(r'\d{1,2}/\d{1,2}/\d{2,4}', wiki)
dates
````
