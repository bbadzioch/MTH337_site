## Project

[Collatz sequences.](../../projects/collatz/collatz.ipynb)

```python
def collatz_list(n):
    
    cl = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        cl.append(n)
    return cl
```
