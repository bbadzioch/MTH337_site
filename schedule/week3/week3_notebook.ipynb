```python
def collatz_list(n):
    
    cl = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        cl.append(n)
    return cl
```

```python
collatz_list(6)
```

```python
import matplotlib.pyplot as plt

plt.plot(collatz_list(6), 'bo-')
plt.show()
```

```python
def get_stops(n):
    stops = [0]
    for x in range(1, n+1):
        cl = collatz_list(x)
        length = len(cl)
        stops.append(length)
    return stops
```

```python
get_stops(10)
```

```python
collatz_list(7)
```
