```python
def collatz_list(n):
    
    cl = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        cl.append(n)
    return cl
```

```python
collatz_list(15)
```

```python
c = collatz_list(31)
```

```python
print(c)
```

```python
len(c)
```

```python
max(c)
```

```python
sum(c)/len(c)
```

```python
plt.figure(figsize=(8, 2))
plt.plot([1, 2, 3], [0, 7, 2], 'b-')
plt.plot([1, 2, 3], [0, 7, 2], 'ro', ms=6)
plt.title("My plot")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.show()
```

```python
m = 15
n = 17
cl1 = collatz_list(m)
cl2 = collatz_list(n)

plt.plot(cl1, 'ro-', label=f"m={m}")
plt.plot(cl2, 'bo-', label=f"n={n}")
plt.legend()
plt.show()
```

```python
m = 15
n = 17
cl1 = collatz_list(m)
cl2 = collatz_list(n)

plt.plot(cl1, 'ro-', label=f"m={m}")
plt.plot(cl2, 'bo-', label=f"n={n}")
plt.xlim(0, 5)
plt.ylim(20, 60)
plt.legend()
plt.show()
```
