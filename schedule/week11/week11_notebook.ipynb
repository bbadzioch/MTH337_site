```python
def get_ngrams(txt, n):
    ngrams = {}
    for i in range(len(txt) - (n+1)):
        ng = txt[i:i+n]
        if ng in ngrams:
            ngrams[ng].append(txt[i+n])
        else: 
            ngrams[ng] = [txt[i+n]]
    return ngrams
    
```

```python
import random
from time import sleep

def markov_writer(prompt, ngrams, num_sentences=1):
    print(prompt)
    n = len(next(iter(ngrams)))
    context = prompt[-n:]

    ns = 0
    while True:
        sleep(0.05)
        if context not in ngrams:
            break
        else:
            letter = random.choice(ngrams[context])
            print(letter, end="")
            context = context[1:] + letter
            if letter in ".?!":
                ns += 1
                if ns == num_sentences:
                    break
```
