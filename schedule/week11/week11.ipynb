## Generative Markov Chains

````python
import random
from time import sleep

def markov_writer(prompt, ngrams, num_sentences=1):
    print(prompt)
    n = len(next(iter(ngrams)))
    context = prompt[-n:]

    ns = 0
    while True:
        sleep(0.05)
        if context not in ngrams:
            break
        else:
            letter = random.choice(ngrams[context])
            print(letter, end="")
            context = context[1:] + letter
            if letter in ".?!":
                ns += 1
                if ns == num_sentences:
                    break
````
