```python
def clean_text(txt):
    txt = txt.lower()
    clist = []
    for c in txt:
        if c in string.ascii_lowercase:
            clist.append(c)
        else: 
            clist.append(" ")
    ctxt = "".join(clist)
    return ctxt

s = "This is a test. 1234! Does it work?"
clean_text(s)
    
```
