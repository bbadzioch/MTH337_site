```python
import requests

book_url = "https://www.gutenberg.org/cache/epub/5658/pg5658.txt"

response = requests.get(book_url)

text = response.text
```

```python
print(text[:2000])
```

```python
import string

def clean_text(txt):
    
    letters = []
    for c in txt.lower():
        if c in string.ascii_lowercase:
            letters.append(c)
        else:
            letters.append(" ")

    ctxt = "".join(letters)
    return ctxt


print(clean_text(text)[:2000])
```

```python
"test".isalpha()
```

```python
"test1".isalpha()
```

```python
def clean_text(txt):
    
    letters = []
    for c in txt.lower():
        if c.isalpha():
            letters.append(c)
        else:
            letters.append(" ")

    ctxt = "".join(letters)
    return ctxt
```

```python
print(clean_text(text)[:2000])
```

```python
ctxt = clean_text(text)

words = ctxt.split()
words[:20]
```

```python
word_dict = {}
for w in words:
    if w not in word_dict:
        word_dict[w] = 1
    else: 
        word_dict[w] += 1
```

```python
word_dict
```

```python
word_dict["and"]
```

```python
word_counts = list(word_dict.items())

word_counts[:50]
```
