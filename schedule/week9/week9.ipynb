### Exercise 1

Write a function `clean_text(txt)` that takes as its argument a string `txt` are returns a string `ctxt` such that:

- `ctxt` converts all upper case letters of `txt` to lower case.
- all non-letter characters in `txt` are replaced by the space character `" "` in `ctxt`.

````python
# remove-cell

import string 

def clean_text(txt):
    
    letters = []
    for c in txt.lower():
        if c in string.ascii_lowercase:
            letters.append(c)
        else:
            letters.append(" ")

    ctxt = "".join(letters)
    return ctxt
````

````python
s = "This is a test. 1234! Does it work?"
print(clean_text(s))
````
