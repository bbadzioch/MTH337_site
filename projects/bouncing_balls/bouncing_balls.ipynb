# Bouncing balls

The goal of this project is to study the motion of bouincing balls. In physics this motion is described as 
*inelastic collision* in which the ball looses some fraction of its energy at each bounce. This is described by the equation:
$$
E_n = C^n E_0
$$
where $E_0$ is the initial energy of the ball, $E_n$ is the energy after $n$ bunces, and $0 \leq C < 1$ is a constant that depends 
on properties of the ball and the surface on which it is dropped. This formula can be used to predict times between consecutive 
bounces: if $t_i$ is the duration of time between bounces $i$ and $i+1$ then $t_{i+1} = \sqrt{C} t_i$.
