# Collatz sequences

Let $n_1 \geq 1$ be an integer. We can make a new integer $n_2$ as follows:  

- if $n_1$ is even, then $n_2 = \frac{n_1}{2}$
- if $n_1$ is odd, then $n_2 = 3n_1 + 1$ 

Applying this procedure recursively, we obtain a sequence of integers $n_1, n_2, n_3, \dots $.
The Collatz conjecture says, that no matter what the first integer $n_1$ is, one of the integers in this sequence will be always 1. After that the sequence will start repeating: 
$$
n_1 , \dots, 1, 4, 2, 1, 4, 2, 1, \dots
$$
It is not known if this conjecture is true or not, but the goal of this project is to investigate sequences obtained in this way and make some observations about them. 

Given a sequence $n_1, n_2, \dots, n_k$ obtained as above, we will call it a *Collatz sequence* if $n_k = 1 $, and  it is the first occurence of 1 in this sequence.

**Exercise 1.** Write a function `collatz()`, which for a given integer `n` the Collatz sequence that starts with `n`.

```python
# remove-cell

def collatz(n: int) -> list: 
    """
    Generate a Collatz path starting with 
    an integer n
    """
    trace = [n]
    while True: 
        if n == 1:
            break
        if n%2:
            n = 3 * n + 1
        else: 
            n = n // 2
        trace.append(n)
    return trace
```

```python
# remove-cell

def get_lengths(vals): 
    """
    Get lengths of Collatz Paths df
    """
    if type(vals) == int:
        vals = range(1, vals + 1)
    lengths = [0]
    for i in vals:
        k = len(collatz(i))
        lengths.append(k)
    return lengths

N = 10**5
clengths = get_lengths(N)
```
